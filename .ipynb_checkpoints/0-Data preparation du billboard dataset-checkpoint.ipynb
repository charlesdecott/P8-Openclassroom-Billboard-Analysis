### IMPORT LIB

```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

### IMPORT BILLBOARD DATASET

```python
siteURL = 'http://www.billboard.com/charts/hot-100/' # add year-month-day
```

```python
billboard = pd.read_csv("data/billboard.csv")
```

```python
billboard['date']=billboard['WeekID'].apply(lambda x:datetime(int(x.split('/')[2]), int(x.split('/')[0]), int(x.split('/')[1])))
```

```python
billboard['year']=billboard['date'].apply(lambda x : x.year)
```

```python
billboard['month']=billboard['WeekID'].apply(lambda x:x.split('/')[0])
```

```python
billboard.head()
```

### Les mois avec plus de jour ont plus de chance detre plus representer dans le classement car le classement sort tous les dimanches

```python
billboard['month'].value_counts()
```

# ARTIST DATAFRAME

```python
artists_df=pd.DataFrame(data=billboard['Performer'].unique(),columns=['Performer'])
artists_df.head()
```

### Ajout list of songs

```python
artists_df['songs']=artists_df['Performer'].apply(lambda x:billboard[billboard['Performer'] == x]['Song'].unique())
```

### Ajout : nombre de musique apparau dans le classement

```python
artists_df['numSong'] = artists_df['songs'].apply(lambda x:len(x))
```

```python
artists_df.sort_values(by='numSong', ascending=False).head()
```

### Ajout du nombre de semaine total pour lartiste dans le classement

```python
artists_df['len_weeks']=artists_df['Performer'].apply(lambda x:len(billboard[billboard['Performer'] == x]))
```

### Ajouter année min et max

```python
artists_df['year_min']=artists_df['Performer'].apply(lambda x:int(billboard[billboard['Performer']==x]['year'].min()))
artists_df['year_max']=artists_df['Performer'].apply(lambda x:int(billboard[billboard['Performer']==x]['year'].max()))
```

```python
xx=artists_df['year_min'].value_counts().sort_index()

plt.plot(xx.index,xx.values)
plt.bar(xx.index,xx.values,color='y')
plt.title('Nombre de nouvel artiste par année')
plt.ylabel('Effectif')
plt.show()
```

```python
xx=artists_df['year_max'].value_counts().sort_index()

plt.plot(xx.index,xx.values)
plt.bar(xx.index,xx.values,color='y')
plt.title('Nombre de derniere apparition dun artiste par année')
plt.ylabel('Effectif')
plt.show()
```

### Ajout : Nombre d'année differentes dans le top 100

```python
artists_df['len_year_different']=artists_df['Performer'].apply(lambda x:len(billboard[billboard['Performer']==x]['year'].unique()))
```

### Ajout : Ecart d'année entre la premiere et derniere fois dans le classement

```python
artists_df['year_space']=artists_df['year_max']-artists_df['year_min']+1
```

```python
minxRange,maxxRange=0,30

xx=artists_df[artists_df['year_space']>=minxRange]
xx=xx[xx['year_space']<=maxxRange]['year_space'].value_counts().sort_index()

plt.bar(xx.index,xx.values,color='r')
plt.title('Repartition des longueurs de carriere des artistes')
plt.ylabel('Effectif')
plt.show()
```

### Compute nombre de semaine au top 1 , top 5, top 10, top 50

```python
artists_df['top1']=artists_df['Performer'].apply(lambda x:len(billboard[billboard['Performer']==x][billboard['Week Position']==1]))
```

```python
artists_df['top5']=artists_df['Performer'].apply(lambda x:len(billboard[billboard['Performer']==x][billboard['Week Position']<=5]))
```

```python
artists_df['top10']=artists_df['Performer'].apply(lambda x:len(billboard[billboard['Performer']==x][billboard['Week Position']<=10]))
```

```python
artists_df.sort_values(by='len_weeks', ascending=False)
```

### Regarder le nombre de featuring 

```python
artists_df['is_featuring']=artists_df['Performer'].apply(lambda x: True if 'Featuring ' in x or 'featuring ' in x else False)
```

```python
artists_df['is_featuring'].value_counts()
```

```python
artists_df[artists_df['is_featuring']== True].sort_values(by='top1', ascending=False).head()
```

# SONGS DATAFRAME

```python
songs_df=pd.DataFrame(data=billboard['SongID'].unique(),columns=['SongID'])
```

```python
songs_df
```

```python
for i in songs_df['SongID']:
    tempDF=billboard[billboard['SongID'] == i]
    maxWeek = tempDF['Weeks on Chart'].max()
    maxInstance = tempDF['Instance'].max()
    maxPos = tempDF['Week Position'].min()
    countPos = tempDF['Week Position'].value_counts() 
    
    print(maxWeek,maxInstance,maxPos)
    print(countPos)
    
    break
```

### Nombre de musique sortis par année

```python
songs_df['year']=songs_df['SongID'].apply(lambda x:int(billboard[billboard['SongID']==x]['year'].min()))
```

```python
xx=songs_df['year'].value_counts().sort_index()

plt.plot(xx.index,xx.values)
plt.bar(xx.index,xx.values,color='y')
plt.title('Nombre de musique sortis par année')
plt.ylabel('Effectif')
plt.show()
```

### Compute le nombre de semaine que le titre apparait dans le classment

```python
songs_df['comboWeek']=songs_df['SongID'].apply(lambda x:billboard[billboard['SongID'] == x]['Weeks on Chart'].max())
```

```python
songs_df.sort_values(by='comboWeek', ascending=False)
```

```python
xx=songs_df['comboWeek'].value_counts().sort_index()

plt.plot(xx.index,xx.values)
plt.bar(xx.index,xx.values,color='y')
plt.title('Nombre de semaine dans le classment')
plt.ylabel('Effectif')
plt.show()
```

### Compute le nombre de fois ou la musique est rentré dans le classement (Instance)

```python
songs_df['comboInstance']=songs_df['SongID'].apply(lambda x:billboard[billboard['SongID'] == x]['Instance'].max())
```

```python
songs_df.sort_values(by='comboInstance', ascending=False)
```

```python
songs_df['comboInstance'].value_counts()
```
