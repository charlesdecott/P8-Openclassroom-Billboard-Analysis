### Exemple de 'All I Want For Christmas Is You' de Mariah Carey

**La musique est sortis en 2000**, est rentree une semaine dans le top 100.
Quelque annes plus tard en 2012 la musique rentre a nouveau dans le classement (au periode de noel)
Toutes les années suivantes la musique va a nouveau etre dans le classement aux semaines proches de noel.

**L'apparition en 2012** est surement du a l'apparition des platformes de streaming 

```python
songId='All I Want For Christmas Is YouMariah Carey'

xx=billboard[billboard['SongID']==songId]
xx=xx.sort_values(by='date', ascending=True)
xx
```

```python
import matplotlib.pyplot as plt
from matplotlib.dates import (YEARLY,MONTHLY, DateFormatter,
                              rrulewrapper, RRuleLocator, drange)
import numpy as np
import datetime

# tick every 5th easter
rule = rrulewrapper(MONTHLY, interval=2)
loc = RRuleLocator(rule)
formatter = DateFormatter('%m/%y')

xxx=xx[xx['year'] > 2001]
dates = xxx['date'] #drange(date1, date2, delta)
s = xxx['Week Position']  # make up some random y values


fig, ax = plt.subplots(figsize=(10, 5))
plt.plot_date(dates, s)
ax.xaxis.set_major_locator(loc)
ax.xaxis.set_major_formatter(formatter)
ax.xaxis.set_tick_params(rotation=45, labelsize=8)

plt.show()
```

```python
print("Nombre de fois present dans le classement :",len(xx))

print("\nEffectif par mois :")
print(xx['month'].value_counts())

print("\nEffectif par ans :")
print(xx['year'].value_counts().sort_index())

print("\nMoyenne position :")
print(xx['Week Position'].mean())
```
